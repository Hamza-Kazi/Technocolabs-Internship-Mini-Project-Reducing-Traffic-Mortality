Principle Component Analysis (PCA)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
from sklearn.decomposition import PCA
```

```python
X = data[['perc_fatl_speed','perc_fatl_alcohol','perc_fatl_1st_time']]
Y = data['State']
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split( X, Y, test_size=1/7.0, random_state=0)
```

```python
X_train = sc.fit_transform(train_X)
X_test = sc.transform(test_X)

pca = PCA(n_components=2)
X_pca = pca.fit(X_train)
```

```python
plt.bar(range(1, X_pca.n_components_ + 1),  pca.explained_variance_ratio_)
plt.xlabel('Principal component')
plt.ylabel('Proportion of variance explained')
plt.xticks([1, 2])

evr = pca.explained_variance_ratio_[0].cumsum()[0] + pca.explained_variance_ratio_[1].cumsum()[0]
print("The cumulative variance of the first two principal components is {}\n".format(
    round(evr, 3)))
```

```python
X_pca = pca.fit_transform(X_train)

p_comp1 = X_pca[:, 0]
p_comp2 = X_pca[:, 1]

plt.scatter(p_comp1, p_comp2)
```

```python
from sklearn.cluster import KMeans
```

K_Means Clustering

```python
km = KMeans(n_clusters = 2, random_state = 8)

km.fit(X_train)
 
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = km.labels_)
```

Decision: The divison of the data into two clusters show that the data points in cluster 1 (Purple) have positive variance, which means that the number of accidents are more in this cluster as compared to other cluster (Yellow).
